```python
import cv2
import os
import xml.etree.ElementTree as ET
import random
import pandas as pd
import glob
```

```python
from tqdm import tqdm
```

```python
root_path = 'D://vehicle data//New folder//vehicle-orientation-3'
label_path =  'D://vehicle data//New folder//part_3_labels'
# save_path = 'D://person class//Annoted_images'
dir_list = os.listdir(root_path)
files_jpg = [x.strip('.jpg') for x in dir_list if x.endswith(".jpg")]
```

```python
df = pd.DataFrame(columns=['Image_name','Class_name','x_min','y_min','x_max','y_max','Height','Width'])
for i in tqdm(files_jpg):
    txt_path = os.path.join(label_path,i+'.txt')
    file_path = os.path.join(root_path,i+'.jpg')
    lst = []
    with open(txt_path, "r") as text_file:
        for k,text_line in enumerate(text_file):
            img = cv2.imread(file_path)
            r,c,ch = img.shape
            shape_list = [1,c,r,c,r]
            text_contents = text_line.replace('\n','').split(' ')
            text_contents = [eval(q) for q in text_contents]
            p = []
            for i1, i2 in zip(text_contents, shape_list):
                p.append(i1*i2)
            xmin = int(p[1] - (p[3]/2))
            ymin = int(p[2] - (p[4]/2))
            xmax = int(p[1] + (p[3]/2))
            ymax = int(p[2] + (p[4]/2))
            df2 = pd.DataFrame([[i,p[0],xmin,ymin,xmax,ymax,r,c]],columns=df.columns)
            df = pd.concat([df,df2],ignore_index=True)
            
```
